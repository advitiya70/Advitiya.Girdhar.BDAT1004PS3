# Question 1
# Occupations


```python
# Step 1. Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Step 2. Import the dataset from this address.
# Step 3. Assign it to a variable called users
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|')
```

```python
users
```

```python
# Step 4. Discover what is the mean age per occupation
users.groupby('occupation')['age'].mean()
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least 
users['Males'] = users.gender.apply(lambda x: True if x == 'M' else False)
```

```python
(users.groupby('occupation').Males.sum() / users.groupby('occupation').gender.count()).sort_values(ascending = False)
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages
print('Minimum age for each occupation \n')
print(users.groupby('occupation').age.min())
print('\nMaximum age for each occupation\n')
print(users.groupby('occupation').age.max())
```

```python
# Step 7. For each combination of occupation and sex, calculate the mean age
users.groupby(['occupation' , 'gender']).age.mean()
```

```python
# Step 8. For each occupation present the percentage of women and men
GC = users.groupby(['occupation' , 'gender']).gender.count()
GC
print('The percentage of women and men for each occupation is:')
Occupation_Count = users.groupby('occupation').gender.count()
GC*100/Occupation_Count
```

```python
# Step 14. How much was the revenue for the period in the dataset?
print('The revenue for the period in the dataset:',chipo['item_price'].sum())
```

```python
# Step 16. What is the average revenue amount per order?
print('The average revenue amount per order is:',chipo.groupby('order_id').item_price.agg('mean').mean())
```

```python
# Step 17. How many different items are sold?
print('The number of different items sold:',chipo.item_name.nunique())
```

# Question 7
Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import numpy as np
import matplotlib.pyplot as plt
marriage = [709000, 1667000, 2315000]    
divorce = [56000, 385000, 944000]
X = np.arange(len(marriage))
plt.bar(X, marriage, color = 'Blue', width = 0.25)
plt.bar(X + 0.25, divorce, color = 'Orange', width = 0.25)
plt.legend(['Marriage', 'Divorce'])
plt.xticks([i + 0.25 for i in range(3)], ['1900', '1950', '2000'])
plt.title("Number of marriages and divorces per capita in the U.S.\n")
plt.xlabel('Years')
plt.ylabel('Number of marriages and divorces ')
plt.show()
```

# Question 9
Create a pie chart showing the fraction of all Roman Emperors that were assassinated.

Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
import matplotlib.pyplot as plt
import pandas as pd
roman_emperors= pd.read_csv(r"C:\Users\Mani\Desktop\Sem 1\Data Programming\Assignments\PS3\roman-emperor-reigns.csv") 
roman_emperors.groupby(['Cause_of_Death']).count().plot(kind='pie', y= 'Emperor',legend= False, startangle=90,figsize=(20,12), autopct='%1.1f%%',ylabel='')
plt.title("Causes of Death of Roman Emperors\n")
plt.show()
```
